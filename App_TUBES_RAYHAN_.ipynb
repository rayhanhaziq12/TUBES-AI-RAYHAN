```python
%%writefile app.py

import streamlit as st
import joblib
import pandas as pd

model = joblib.load('best_model.joblib')

st.set_page_config(
    page_title="Flight Ticket Price Prediction",
    layout="centered"
)

st.title("‚úàÔ∏è Flight Ticket Price Prediction")

st.write(
    "Enter flight information to predict ticket price."
)

airline = st.number_input("Airline", min_value=0)
source_city = st.number_input("Source City", min_value=0)
destination_city = st.number_input("Destination City", min_value=0)
duration = st.number_input("Duration (minutes)", min_value=0)

if st.button("Predict"):

    input_data = pd.DataFrame(
        [[airline,
          source_city,
          destination_city,
          duration]],
        columns=[
            'airline',
            'source_city',
            'destination_city',
            'duration'
        ]
    )

    prediction = model.predict(input_data)

    st.success(
        f"Predicted Ticket Price: ‚Çπ {prediction[0]:,.2f}"
    )

st.write("Model R¬≤ Score: 0.97")
```
